```python
import csv
import pandas as pd
import numpy as np
import datetime as dt
import sys
```

```python
filepath="C:\\Users\\user\\Desktop\\data\\"
files=['CO','NO2','O3','PM10','PM25','SO2']
```

```python
def data_cleaning (filename):
    data=pd.DataFrame.from_csv(filepath+filename+'.csv',encoding="utf-8")
    print('1. ',filename,'open 완료')
    data=data.interpolate(method='values', limit=6)
    column=data.columns
    for i in range(len(data)):
        data.iloc[i]=data.iloc[i].fillna(data.iloc[i].mean())
    print('2. ',filename,'cleaning 완료')    
    data.to_csv(filepath+'cleaning\\'+filename+'.csv',encoding="utf-8")
    print('3. ',filename,'write 완료')
    print(data.isnull().sum())
    
```

```python
for file in files :
    data_cleaning(file)
```
